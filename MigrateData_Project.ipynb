```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("เมนูอาหาร - เมนูอาหาร Update.csv")
df
```

```python
df["Calories(Kcal)"] = df["Calories(Kcal)"].astype(float)
df["Carbohydrate(g)"] = df["Carbohydrate(g)"].astype(float)
df["Fat(g)"] = df["Fat(g)"].astype(float)
df["Protein(g)"] = df["Protein(g)"].astype(float)
df["Sodium(mg)"] = pd.to_numeric(df["Sodium(mg)"], errors='coerce')
df["Sodium(mg)"] = df["Sodium(mg)"].astype(float)
df["Sugar(g)"] = df["Sugar(g)"].astype(float)
df.dtypes
```

```python
df["menu_img"] = np.nan
```

```python
df.columns
```

```python
df["CookingType"].isna().sum()
```

```python
df["CookingType"].value_counts()
```

```python
df["CookingType"] = df["CookingType"].replace({
    "ต้ม/นึง\nเส้น": "ต้ม/นึ่ง\nเส้น",
    "ต้ม/นึง": "ต้ม/นึ่ง",
    "\nต้ม/นึง\nเส้น": "ต้ม/นึ่ง\nเส้น",
    "ผัด/ทอด\nต้ม/นึ่ง": "ต้ม/นึ่ง\nผัด/ทอด"
})
df["CookingType"].value_counts()
```

```python
df["CookingType"] = df["CookingType"].str.split("\n")
print(df[["CookingType"]].head())
```

```python
cooking_type_map = {
    "ต้ม/นึ่ง": 1,
    "ปิ้ง/ย่าง": 2,
    "ยำ": 3,
    "เส้น": 4,
    "ผัด/ทอด": 5
}
def map_cooking_type_to_id(cooking_types):
    return [cooking_type_map.get(item, 0) for item in cooking_types]

df["cooking_type_id"] = df["CookingType"].apply(lambda x: map_cooking_type_to_id(x) if isinstance(x, list) else [])
df
```

```python
def create_nutrition(row):
    return {
        "calorie": row["Calories(Kcal)"],
        "fat": row["Fat(g)"],
        "carbohydrate": row["Carbohydrate(g)"],
        "protein": row["Protein(g)"],
        "sodium": row["Sodium(mg)"],
        "sugar": row["Sugar(g)"],
        "glycemic_index": 0
    }

df["nutrition"] = df.apply(create_nutrition, axis=1)
print(df[["Calories(Kcal)", "nutrition"]].head())
```

```python
def format_ingredient(ingredient):
    if pd.isna(ingredient):
        return []
    return [{"name": item.strip()} for item in ingredient.split("\n")]

df["ingredients"] = df["Ingredient"].apply(format_ingredient)
df
```

```python
df["quantity"] = "1 จาน/ถ้วย"
```

```python
df["category_id"] = 1
```

```python
df
```

```python
df_2 = df[["Name","menu_img","category_id","CookingType","nutrition","ingredients","ColorType","describe","cooking_type_id","quantity"]].rename(
    columns={"CookingType": "cooking_type",
             "ColorType": "color",
             "Name": "name",
             }
)
```
